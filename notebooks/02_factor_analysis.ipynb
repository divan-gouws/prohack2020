## factor analysis

the gotcha here will be the sequential nature of the data introduced by the galaxy year

```
from factor_analyzer import FactorAnalyzer
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
from factor_analyzer.factor_analyzer import calculate_kmo
```

```
%store -r ngc_1560_df
ngc_1560_df
```

Factor analysis can only be applied to interval data, which is extracted as follows:

```
galaxy_numeric = ngc_1560_df[ngc_1560_df.columns.difference(['galactic year', 'galaxy'])]
```

```
chi_square_value, p_value = calculate_bartlett_sphericity(galaxy_numeric)
chi_square_value, p_value
```

Bartlett’s test of sphericity checks whether or not the observed variables intercorrelate at all using the observed correlation matrix against the identity matrix.

The $p$-value is significant, so the observed correlation matrix is not an identity matrix and this assumption is met for factor analysis.

```
kmo_all, kmo_model = calculate_kmo(galaxy_numeric)
kmo_model
```

Kaiser-Meyer-Olkin (KMO) Test measures the suitability of data for factor analysis. It determines the adequacy for each observed variable and for the complete model. KMO estimates the proportion of variance among all the observed variable.

Values greater than 0.6 indicate that the data is suitable for factor analysis. In this case, the KMO test produces a value of 0.92 which is fine.

```
fa = FactorAnalyzer(rotation='oblimin', n_factors=2)
fa.fit(galaxy_numeric)
ev, v = fa.get_eigenvalues()
ev
```
